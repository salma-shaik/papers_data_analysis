```python
import pandas as pd
import xml.etree.ElementTree as ET

tree = ET.parse("/Users/salma/Research/papers_data_analysis/bda_ml_wm_lr/EBSCO_2020-09-26.xml")

root = tree.getroot()

data = []

for record in tree.findall('.//rec/header/controlInfo'):
    article_dict = {}
    
    article_dict['artTitle'] = record.find('./artinfo/tig/atl').text
    article_dict['pubYear'] = record.find('./pubinfo/dt').text
    article_dict['pubJournal'] = record.find('./jinfo/jtl').text
    article_dict['artAbstract'] = record.find('./artinfo/ab').text
    article_dict['pubType'] = record.find('./artinfo/pubtype').text
  #  article_dict['artKeywords'] = record.find('./artinfo/sug/subj')
    
#     for kw in root.iter('sug'):
    subjs = []
    
    values = record.findall('./artinfo/sug/subj')
    for val in values:
        #print(val.text)
        subjs.append(val.text)
        
    kws_str = "".join(subjs)
    
    article_dict['artKeywords'] = kws_str
    data.append(article_dict)

keywords = []
ebsco_articles = pd.DataFrame(data)
ebsco_articles_req = ebsco_articles.query('pubType == "Academic Journal"').drop('pubType', axis=1)

print(list(ebsco_articles_req))
print(ebsco_articles_req.shape[0])
ebsco_articles_req.to_csv('/Users/salma/Research/papers_data_analysis/bda_ml_wm_lr/db_req_fields_articles/bda_ml_wm_ebsco_req_s1.csv',
                     index=False)
```
